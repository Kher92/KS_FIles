```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_excel('ICP_Ã„rzte.xlsx',engine='openpyxl')
```

```python
df.isnull().sum()
```

```python
df_duplicate = df.duplicated(subset=['Fachrichtung','Name','Ort','Straße'], keep=False)
```

```python
df_a = df[df_duplicate]
```

```python
df_a
```

```python
df_clean= df[~df_duplicate]
```

```python
df_nan = df_clean.isnull()
```

```python
df_nan
```

```python
df_clean= df_clean[~df_nan]
```

```python
df_clean
```

```python
df_clean.columns
```

```python
df_clean = df_clean.rename(columns=lambda colu:colu.replace('/', '').replace('\n', ''))
```

```python
df_clean
```

```python
df_clean.isnull().sum()
```

```python
df_clean = df_clean.rename(columns=lambda column_name: column_name.replace('Airbnb_Open_Data.csv', ''))
```

```python
df_clean
```

```python
df_clean = df_clean[~df_a]
```

```python
df_clean
```

```python
df_clean = df_clean[['Ort','Straße','Fachrichtung','Gesamtkumuliert2025November','Gesamtkumuliert in Stück2025November','Name','ID-Nr']]
```

```python
df_clean
```

```python
df_clean = df_clean.rename(columns={'Gesamtkumuliert2025November':'Wert', 'Gesamtkumuliert in Stück2025November':'Stuck'})
```

```python
df_clean
```

```python
df_top_fachrichtung = df_clean.groupby('Fachrichtung').agg(
    Anzal = ('ID-Nr','count'),
    gesamt_wert = ('Wert','sum'),
    gesamt_stuck=('Stuck','sum')
)
```

```python
df_top_fachrichtung
```

```python
df_top = df_top_fachrichtung.sort_values(by='gesamt_wert', ascending=False).head(10)
```

```python
df_top
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Fachrichtung', y='gesamt_wert', data=df_top, palette='viridis')

plt.title('Top 10 Fachrichtungen nach Gesamtwert')
plt.xlabel('Fachrichtung')
plt.ylabel('Gesamtwert (EUR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_Fachrichtung_barplot.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Fachrichtung', y='gesamt_wert', data=df_top, palette='pastel')

plt.title('Vergleich des Gesamtwerts pro Fachrichtung')
plt.xlabel('Fachrichtung')
plt.ylabel('Gesamtwert (EUR)')
plt.xticks(rotation=45) # Dreht die X-Achsen-Labels, falls sie zu lang sind

plt.tight_layout()
plt.show()
```

```python
df_praxis = df_clean.groupby(['Ort','Name']).agg(
    Anzahl = ('ID-Nr','count'),
    gesamt_wert = ('Wert','sum'),
    gesamt_stuck=('Stuck','sum')
).reset_index()
```

```python
df_top_praxis = df_praxis.sort_values(by='gesamt_wert',ascending=False)
```

```python
df_top_praxis['Praxis_Ort'] = df_top_praxis['Name'] + ' (' + df_top_praxis['Ort'] + ')'
```

```python
df_top_praxis
```

```python
df_top_praxis_top10 = df_top_praxis.head(10)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x="gesamt_wert", y="Praxis_Ort", data=df_top_praxis_top10, palette='viridis')
plt.xlabel('Gesamtwert (EUR)')  
plt.ylabel('Praxis (Name & Ort)') 

plt.tight_layout()
plt.savefig('top_praxis_barplot.png', dpi=300, bbox_inches='tight')
plt.show()

```
